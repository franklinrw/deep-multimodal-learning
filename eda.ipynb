```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load_data(filepath):
    """Load data from a pickle file."""
    with open(filepath, 'rb') as file:
        data = pickle.load(file)
    return data

def visualize_images(images, labels, num_images=5):
    """Visualize a few images from the dataset."""
    for i in range(min(num_images, len(images))):
        image = images[i].reshape(128, 64, 3)  # Reshape to have init and effect images stacked vertically
        init_image = image[:64, :, :]
        effect_image = image[64:, :, :]
        
        # Concatenate the init and effect images horizontally (side by side)
        concatenated_image = np.hstack((init_image, effect_image))
        
        plt.imshow(concatenated_image)
        plt.title(f'Label: {labels[i]}')
        plt.axis('off')
        plt.show()
        
def label_distribution(labels):
    """Plot the distribution of labels in the dataset."""
    plt.hist(labels, bins=np.arange(len(np.unique(labels)) + 1) - 0.5, edgecolor='black')
    plt.xlabel('Label')
    plt.ylabel('Frequency')
    plt.title('Label Distribution')
    plt.xticks(np.arange(len(np.unique(labels))))
    plt.grid(axis='y')
    plt.show()

def image_statistics(images):
    """Calculate and display some statistics about the images in the dataset."""
    print(f'Number of images: {len(images)}')
    print(f'Image shape: {images[0].shape}')
    print(f'Image data type: {images[0].dtype}')
    print(f'Min pixel value: {np.min(images[0])}')
    print(f'Max pixel value: {np.max(images[0])}')
    print(f'Mean pixel value: {np.mean(images[0])}')
    print(f'Standard deviation of pixel values: {np.std(images[0])}')
```

```python
# Example usage:
# Load data
training_color = load_data('data/training_color.pkl')
y_training = load_data('data/y_training.pkl')

# Visualize some images
visualize_images(training_color, y_training)

# Plot label distribution
label_distribution(y_training)

# Display image statistics
image_statistics(training_color)
```
