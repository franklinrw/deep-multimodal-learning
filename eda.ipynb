```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load_data(filepath):
    """Load data from a pickle file."""
    with open(filepath, 'rb') as file:
        data = pickle.load(file)
    return data

def decode_label(label):
    """Decode a label tuple into its corresponding names."""
    tool_label, action_label = label
    return tool_mapping[tool_label], action_mapping[action_label]

def visualize_single_image(image, label):
    decoded_label = decode_label(label)
    plt.imshow(image)
    plt.title(f'Tool: {decoded_label[0]}, Action: {decoded_label[1]}')
    plt.axis('off')
    plt.show()

toolnames = ['hook', 'ruler', 'spatula', 'sshot']
actions = ['left_to_right', 'pull', 'push', 'right_to_left']

tool_mapping = {i: tool for i, tool in enumerate(toolnames)}
action_mapping = {i: action for i, action in enumerate(actions)}
```

```python
filepath = 'C:/Users/Frank/OneDrive/Bureaublad/ARC/deep-multimodal-learning/data/10_tomatoCan/hook/left_to_right/color/training.pkl'
images = load_data(filepath)

label_filepath = 'C:/Users/Frank/OneDrive/Bureaublad/ARC/deep-multimodal-learning/data/10_tomatoCan/hook/left_to_right/color/y_training.pkl'
labels = load_data(label_filepath)
```

```python
image = images[0].reshape(192, 512, 3)
label = labels[0]
```

```python
visualize_single_image(image, label)
```
